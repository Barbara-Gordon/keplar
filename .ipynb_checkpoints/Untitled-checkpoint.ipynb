```python
# Data Catalog https://exoplanetarchive.ipac.caltech.edu/docs/API_kepcandidate_columns.html

import pandas as pd
import numpy as np
from numpy.random import seed
seed(1)

import tensorflow
tensorflow.keras.__version__
```

```python
df = pd.read_csv("cumulative.csv")
df = df.drop(columns=["rowid", "kepid", "kepoi_name", "kepler_name", "koi_pdisposition", "koi_score", "koi_tce_delivname"])
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.info()
```

```python
X = df.drop('koi_disposition', axis =1)
y = df['koi_disposition']
print(X.shape,y.shape)
y.unique()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)


# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
print(encoded_y_test,y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create a Neural Network model here


model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=40))
model.add(Dense(units=6, activation='relu', input_dim=40))
model.add(Dense(units=6, activation='relu', input_dim=40))
model.add(Dense(units=3, activation='softmax'))
model.summary()
```

```python
# Compile and fit the model
# YOUR CODE HERE
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose = 2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
print(f"Predicted class: {model.predict_classes(X_test_scaled[:10])}")
print(f"Actual Labels: {list(encoded_y_test[:10])}")
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled, encoded_y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
predictions = grid.predict(X_test_scaled)
from sklearn.metrics import classification_report
print(classification_report(encoded_y_test, predictions,target_names = ["Candidate","Confirmed","FalsePositive"]))
```

Both the 3 node Neural Network and the GridSearch model were able to predict whether a planet was a Candidate/Confirmed/FalsePostive about 89% of the time using the factors provided.
